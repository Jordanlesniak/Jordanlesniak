### 1.2 Methodology:

The Transit Method is the most popular method for finding exoplanets. The transit method works by observing a star system, and looking for any variation in the amount of star-light (flux) coming from that particular system. The way that the flux can vary for a particular system relies on a couple of things. First and most importantly, it depends on our point-of-view on any star system that we decide to look at. If we are trying to find planets orbiting around a star, if we are looking from an angle that is 90 degrees with respect to the orbital plane, we'll never find a transiting planet because there is a zero percent chance that a planet will pass in front of its host star, from our point of view. So we are looking for systems where, the planets in that system orbit in such a way that from our point of view, we are directly in line with it's orbital plane. This way when a planet orbits its star, at some point during that orbit the planet will 'transit' in front of its host star, from our point of view. In the process, the planet blocks a tiny amount of the flux coming from that star, which we are able to detect. If this signal of dampened flux is periodic, and we can see multiple periods, then without a doubt, we have found an exoplanet! The next most important thing for this method is, the distance at which a planet orbits. Not only will a planet orbiting at a farther distance block less star-light, the signal of dampened star-light, will also be less periodic, which is key for confirming an exoplanet's existence. The last thing that is important for the transit method, is the radius of the planet/star. The bigger the radius of the planet, the more light that is blocked, the larger the signal is, the easier it is to confirm that an object is clearly transiting. Vis Versa, the smaller the radius of the star, the more light a planet will block, the larger the signal is, the easier it is to confirm the exoplanet.
